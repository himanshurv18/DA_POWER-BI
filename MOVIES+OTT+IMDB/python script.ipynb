```python
import pandas as pd

# Path to the CSV file
file_path = 'C:/Users/HIMANSHU/Desktop/Data Analyst Bootcamp/Assignments/PROJECT PORTFOLIO/MOVIES+OTT+IMDB/Netflix.csv'

# Read the CSV file into a DataFrame
data = pd.read_csv(file_path)

# List of columns to keep as is
columns_to_keep = ['type', 'title', 'director', 'cast', 'country', 'date_added', 'release_year', 'rating', 'duration', 'listed_in', 'description', 'day_added', 'year_added', 'month_added', 'User_Rating']

# Melting the DataFrame to combine genre columns into one
melted_data = pd.melt(data, id_vars=columns_to_keep, var_name='genre', value_name='genre_value')

# Filtering out rows where genre_value is 1
melted_data = melted_data[melted_data['genre_value'] == 1]

# Dropping the 'genre_value' column as it's not needed anymore
melted_data = melted_data.drop('genre_value', axis=1)

# Grouping by columns_to_keep and aggregating genres into a single string
grouped_data = melted_data.groupby(columns_to_keep)['genre'].agg(', '.join).reset_index()

# Saving the modified DataFrame to a new CSV file
new_file_path = 'C:/Users/HIMANSHU/Desktop/Data Analyst Bootcamp/Assignments/PROJECT PORTFOLIO/MOVIES+OTT+IMDB/Netflix_New.csv'
grouped_data.to_csv(new_file_path, index=False)
```
